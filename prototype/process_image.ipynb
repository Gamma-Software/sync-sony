```python
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = 5, 10
from matplotlib.pyplot import imshow
import numpy as np
import scipy.signal
import scipy.interpolate
import scipy.stats
import PIL.Image
from PIL import ImageEnhance, ImageFilter, Image
import rawpy
import imageio
```

```python
file = "image.dng"
img = Image.open(file)
```

```python
with rawpy.imread(file) as raw:
    rgb = raw.postprocess(use_camera_wb=True)
# imageio.imsave('default.tiff', rgb)
imshow(rgb)
```

```python
image = PIL.Image.fromarray(rgb)
img2 = image.filter(ImageFilter.SHARPEN)
imshow(img2)
```

```python
applier = ImageEnhance.Contrast(img2)
img = applier.enhance(0.9)
imshow(img)
```

```python
filter = ImageEnhance.Sharpness(img)
img2 = filter.enhance(2)
imshow(img2)
```

```python
filter = ImageEnhance.Color(img2)
img3 = filter.enhance(0.9)
imshow(img3)
```

```python
img3.save('image.jpg', quality=100, subsampling=0)
img3.save('image.tiff')
```
